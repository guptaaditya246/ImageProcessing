```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('images/contour1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image2 = image.copy()
f, ax = plt.subplots(figsize=(10, 10))
ax.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

#converting the color of image to find better contours
retval, binary = cv2.threshold(gray, 230, 255,cv2.THRESH_BINARY_INV)

f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))
ax1.imshow(gray, cmap='gray')
ax1.set_title("Original Gray Image")
ax2.imshow(binary, cmap='gray')
ax2.set_title("Inverted Gray Image")
```

```python
#finding contours
retval, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contour_image = cv2.drawContours(image2, contours, -1, (0,255,0), 3)
plt.imshow(contour_image)
```

```python
#Localizing contours
for contour in contours:
    x,y,w,h = cv2.boundingRect(contour)
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 3)
    plt.imshow(image)
```
