```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
image = cv2.imread('images/corner4.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
image_copy = image.copy()
```

```python
gray  =cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
gray = np.float32(gray)
corner = cv2.cornerHarris(gray, blockSize = 2, ksize = 3, k=0.04)
corner  =cv2.dilate(corner, None)
plt.imshow(corner, cmap='gray')
```

```python
threshold = 0.01*corner.max()

for i in range(0, corner.shape[0]):
    for j in range(0, corner.shape[1]):
        if(corner[i,j] > threshold):
            cv2.circle(image_copy, (j,i), 2, (0,255,0), 1)

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(image_copy)
        
```
