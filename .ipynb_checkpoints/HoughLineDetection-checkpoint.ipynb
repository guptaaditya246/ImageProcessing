```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
image = cv2.imread('mosque.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
low = 90
high = 100
edges = cv2.Canny(gray, low, high)
plt.imshow(edges)
image_copy = image.copy()
```

```python
rho = 1
theta = np.pi/180
threshold = 50
min_lin_length = 180
max_line_gap = 5

lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), min_lin_length, max_line_gap)
for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(image_copy, (x1, y1), (x2, y2), (0, 255, 0), 5)
        
plt.imshow(image_copy)
```
